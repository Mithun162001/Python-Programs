<h1>Data visualization

```python
import matplotlib.pyplot as plt    # (or) from matplotlib import pyplot as plt
%matplotlib inline                 # magic command to plot in the same window
```

```python
fig = plt.figure()     

# creates a figure with 0 axes
```

```python
# to plot a straight lines

plt.plot([1,1])    # (x,y)
```

```python
# to plot multiple straight lines

plt.plot([1,1])
plt.plot([2,2])
plt.plot([3,3])
```

```python
# Label:- is used to add labels for respectives x and y axis
# title:- is used to display the title of the plot

plt.plot([1,2,3,4])
plt.title("Demo title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Color abbrevations:-

* blue - b
* green - g
* red - r
* cyan - c
* black - k
* white - w

```python
plt.plot([1,2,1,2], 'k')
plt.title("Color abbrevation example")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
# subplots:- to plot multiple axes in single figure
# (nrows,ncols,index_plot)

plt.subplot(2,2,1)
plt.plot([1,4], color='green')

plt.subplot(2,2,2)
plt.plot([2,2], color='red')

plt.subplot(2,2,3)
plt.plot([2,2], color='red')

plt.subplot(2,2,4)
plt.plot([2,2], color='red')
```

```python
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]

plt.plot(x1,y1, color='red', label = 'line one', linewidth = 1, marker='h')
plt.plot(x2,y2, color='blue', label = 'line two', linewidth = 5)
plt.legend()
```

Types of plots:-
* Bar graph
* Histogram
* Scatter plot
* Area plot
* Pie chart

bar plot: A bar graph is the graphical representation of categorical data using rectangular bars where the length of each bar is proportional to the value they represent

```python
import numpy as np
usn = np.arange(5,51,5)
Lab = [90,89,78,79,80,90,92,95,89,99]
theory=[10,20,30,40,50,60,70,80,90,89]
fig = plt.figure(figsize=(10,5))
plt.bar(usn,Lab,label='lab')
plt.legend()

plt.xlabel("USN")
plt.ylabel("Marks")
plt.title("Marks of students")
```

```python
plt.barh(usn,theory,label='theory')
plt.legend()

plt.xlabel("USN")
plt.ylabel("Marks")
plt.title("Marks of students")
```

A histogram is the graphical representation of data where data is grouped into continuous number ranges and each range corresponds to a vertical bar.

```python
age = [22,32,34,56,87,77,55,90]
bins = np.arange(0,101,10)

plt.hist(age, bins, histtype='step', rwidth = 1, color= 'black')
```
