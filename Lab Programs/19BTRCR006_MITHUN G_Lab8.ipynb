```python
# importing the libraries

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("F:\Datasets\loan_data_set.csv")
df.head()
```

```python
# finding the number of missing values present in the data
print("No. of missing values present in the data:\n\n",df.isna().sum())
```

```python
# total number of missing values present in the data
print("Total number of missing values present: ",df.isna().sum().sum())
```

```python
# dropping rows of a dataframe if they have NA values if how='any'
dropped_df = df.dropna(how='any')
print("Total number of missing values after dropping: ",dropped_df.isna().sum().sum())
```

```python
# dropping rows of a dataframe if they have NA values if how='all'
dropped_df2 = df.dropna(how='all')
print("Total number of missing values after dropping: ",dropped_df2.isna().sum().sum())

# this is because there are no  NA values present in a single row
```

```python
# bfill:- the current NA cells will be filled from the value present in the next column in the same row. 
# If the next column is also NA cell then it wonâ€™t be filled.

df1 = df.fillna(method='bfill')
print("Total number of missing values after dropping: ",df1.isna().sum().sum())
print()
df1[:5]
```

```python
# ffill():- function to fill the missing values along the column axis. 
# When ffill is applied across the column axis, then missing values are filled by the value in previous column in the same row.
```

```python
df2 = df.fillna(method='ffill')
print("Total number of missing values after dropping: ",df2.isna().sum().sum())
print()
df2[:5]

#  values in the first row is still NaN value because there is no row above it from which non-NA value could be propagated.
```

```python
df_mean = df.fillna(df.mean())
print("Total number of missing values after dropping: ",df_mean.isna().sum().sum())
print()
df_mean[:5]  

# this is because mean cannot be calculated for the categorical values
```
